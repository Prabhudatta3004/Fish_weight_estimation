##We can see that two fishes in our dataset have Length1 way different from the others. So, these two values are outliers for Length1 in our dataset.

```python
##Checking for Outliners
dataset_L1 = dataset['Length1']

#quartiles
dataset_L1_q1 = dataset_L1.quantile(0.25)
dataset_L1_q2 = dataset_L1.quantile(0.5)
dataset_L1_q3 = dataset_L1.quantile(0.75)

#inter quartile range
dataset_L1_iqr = dataset_L1_q3 - dataset_L1_q1

#lower whisker point
dataset_L1_lower = dataset_L1_q1 - (1.5 * dataset_L1_iqr)

#upper whisker point
dataset_L1_upper = dataset_L1_q3 + (1.5 * dataset_L1_iqr)

dataset_L1_outliers = dataset_L1[(dataset_L1 < dataset_L1_lower) | (dataset_L1 > dataset_L1_upper)]
dataset_L1_outliers
```

```python
##so length1 having values 142,143,144 are outliners
```

# Machine Learning Model

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=dataset.iloc[:,2:]##independent variables
y=dataset.iloc[:,1]##dependent variable
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print("Shape of X_train: ", x_train.shape)
print("Shape of y_train: ", x_test.shape)
print("Shape of X_test: ", y_train.shape)
print("Shape of y_test: ", y_test.shape)
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
print("Intercept of the model: ", regressor.intercept_)
print("Coefficient of the model: ", regressor.coef_)
```

## PREDICTION

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
```

```python
print(score)
```

```python
##i.e the model is 89% accurate approx
```
